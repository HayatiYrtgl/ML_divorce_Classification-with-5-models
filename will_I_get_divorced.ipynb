### TRAIN-TEST-SPLIT & PREPROCESSING

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score, classification_report, confusion_matrix, accuracy_score
from keras.layers import *
from keras.models import Sequential
```

```python
# copy
preprocessed_data = data.copy()
```

```python
# train test split
y = preprocessed_data.Divorce.values
X = preprocessed_data.drop("Divorce", axis=1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
X_train.shape, y_train.shape
```

### EVALUATE FUNCTION

```python
scores = dict()
def evaluate_model(model, test_data, true_data, model_name: str):
    predicted_values = model.predict(test_data)
    accuracy = accuracy_score(true_data, predicted_values)
    f1 = f1_score(y_true=true_data, y_pred=predicted_values)
    scores[model_name] = [accuracy, f1]
    #print(classification_report(y_pred=predicted_values, y_true=true_data))
    #print(confusion_matrix(y_true=true_data, y_pred=predicted_values))
```

### LOGISTIC REGRESSION

```python
logistic_regression = LogisticRegression(solver="liblinear", penalty="l2")
logistic_regression.fit(X_train, y_train)
```

```python
evaluate_model(logistic_regression, X_test, y_test, "logistic_regression")
```

```python
logistic_regression.predict_proba(X_test)
```

### Save logistic model

```python
import joblib
```

```python
joblib.dump(logistic_regression, "divorce_models/logistic_regression_model.pkgl")
```

### KNN CLASSIFIER

```python
knn = KNeighborsClassifier(n_jobs=4, n_neighbors=2)
knn.fit(X_train, y_train)
```

```python
evaluate_model(knn, X_test, y_test, "KNN-classifier")
```

```python
knn.predict_proba(X_test)
```

```python
joblib.dump(knn, "divorce_models/kneigborsClassifier.pkl")
```

### Random Forests

```python
rf = RandomForestClassifier(n_estimators=120, max_depth=10, n_jobs=4)
rf.fit(X_train, y_train)
```

```python
evaluate_model(rf, X_test, y_test, "randomForest-classifier")
```

```python
joblib.dump(rf, "divorce_models/random_forests_classfier.pkl")
```

### Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
evaluate_model(gnb, X_test, y_test, "GaussianNB")
```

```python
joblib.dump(gnb, "divorce_models/gaussian_nb.pkl")
```

###  ANN

```python
model = Sequential()
model.add(Dense(54, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(54, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(54, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(54, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(x=X_train, y=y_train, epochs=25, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
acc_ann = model.evaluate(x=X_test, y=y_test)[1]
```

```python
scores["ANN"] = [acc_ann, np.nan]
```
